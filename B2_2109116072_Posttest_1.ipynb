## Algoritma Naive Bayes dan Preprocessing (pengenalan)

```python
# sebelum membuat algoritma pertama tama buatlah dataset baru dengan ketentuan
# mengambil dataset df
# Menghapus kolom Unnamed: 0, CLIENTNUM, Umur, dan Limit Kartu Kredit (initinya yang berupa primary key dan bukan tipe kategori)
# dataset dengan kolom yang sudah dihapus tadi kemudian disimpan di dataframe baru (df baru)
```

```python
# Buatlah sebuah fungsi naive bayes dengan ketentuan
# Mengambil data dari dataframe yang telah diproses pada blok sebelumnya (diatas)
# Nama fungsinya adalah naive_bayes
# Mengambil 1 saja parameter yaitu sebuah list berisi kategori

# Outputnya berupa prediksi Attrited Customer atau Existing Customer
# Outputnya berupa tuple (_hasilnya, persentasenya berapa) contoh (Existing Customer, 0.004)

# Tidak boleh menggunakan library manapun selain pandas dan numpy



# contoh
# def naive_bayes(prediksi:list)-> tuple :
#   ........

# jadi nanti saat saya coba masukan data 
# naive_bayes(["F",	"Uneducated","Single", "Unknown	Blue"]) akan keluar tuple berupa ("Attired Customer", 0.002)
```

```python
def naive_bayes(kategori: list)-> tuple :
    df = pd.read_csv("ChurnBank.csv")
    df.pop("Unnamed: 0")
    df.pop("CLIENTNUM")
    df.pop("Customer_Age")
    df.pop("Credit_Limit")
    dfbaru = df
    # filter dataset berdasarkan kategori
    dfbaru = dfbaru.loc[dfbaru["Card_Category"].isin(kategori)]

    # Menghitung probabilitas Attrited Customer dan Existing Customer
    total = len(dfbaru)
    attrited = len(dfbaru[dfbaru["Attrition_Flag"] == "Attrited Customer"])
    existing = total - attrited
    attrited_prob = attrited / total
    existing_prob = existing / total

    # Menghitung probabilitas bersyarat untuk setiap fitur
    features = ["Gender", "Education_Level", "Marital_Status", "Income_Category"]
    probs = {}
    for feature in features:
        probs[feature] = {}
        for value in dfbaru[feature].unique():
            count = len(dfbaru[dfbaru[feature] == value])
            attrited_count = len(dfbaru[(dfbaru[feature] == value) & (dfbaru["Attrition_Flag"] == "Attrited Customer")])
            existing_count = count - attrited_count
            probs[feature][value] = {
                "Attrited Customer": attrited_count / attrited,
                "Existing Customer": existing_count / existing
            }

    # Membuat prediksi berdasarkan input
    input_data = {
        "Gender": kategori[0],
        "Education_Level": kategori[1],
        "Marital_Status": kategori[2],
        "Income_Category": kategori[3],
    }
    attrited_prob_input = attrited_prob
    existing_prob_input = existing_prob
    for feature, value in input_data.items():
        attrited_prob_input *= probs[feature][value]["Attrited Customer"]
        existing_prob_input *= probs[feature][value]["Existing Customer"]
    if attrited_prob_input > existing_prob_input:
        result = ("Attrited Customer", attrited_prob_input)
    else:
        result = ("Existing Customer", existing_prob_input)

    # Hasil prediksi
    return tuple(result)
```
